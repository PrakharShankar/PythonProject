```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('State wise literacy rates.csv')
for i in range(1951 , 2012 , 10):
    plt.figure(figsize = (14 , 8))
    plt.bar(data['State/ UTs'] , data[str(i)] , color = 'skyblue')
    plt.xlabel('states')
    plt.ylabel(str(i))
    plt.xticks(rotation = 90)
    plt.ylim(0 , 100)
    plt.show()
```

```python
import numpy as np
data_new = data[['State/ UTs' , '1951' , '1961' , '1971' , '1981']]
x = np.arange(len(data_new['1951']))
width = 0.25
plt.figure(figsize = (12 , 8))
plt.bar(x - width , data_new['1951'], width , label = '1951' , color = 'cadetblue')
plt.bar(x , data_new['1961'], width , label = '1961' , color = 'burlywood')
plt.bar(x + width , data_new['1971'], width , label = '1971' , color = 'darkmagenta')
plt.xticks(x , data_new['State/ UTs'] , rotation = 90)
plt.xlabel("states/UTs")
plt.ylabel("literacy rates(%)")
plt.ylim(0 , 100)
plt.legend()
plt.show()
```
