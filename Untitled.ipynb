```python
import pdfplumber
import pandas as pd

def extract_table_from_pdf(input_pdf_path):
    with pdfplumber.open(input_pdf_path) as pdf:
        # Assuming the table is on the first page
        first_page = pdf.pages[0]
        table = first_page.extract_table()
        return table

def save_to_csv(table, output_csv_path):
    df = pd.DataFrame(table[1:], columns=table[0])
    df.to_csv(output_csv_path, index=False)

# Specify the input PDF file path and the output CSV file path
input_pdf_path = 'sdp current prices.pdf'
output_csv_path = 'output.csv'

# Extract table from PDF
table_data = extract_table_from_pdf(input_pdf_path)

# Save table as CSV
save_to_csv(table_data, output_csv_path)

print(f"Conversion successful! Data saved as '{output_csv_path}'.")
```

```python
import pandas as pd

# Load the Excel file
xls = pd.ExcelFile('final GDP.xlsx')

# Get the sheet names
sheet_names = xls.sheet_names

# Initialize an empty DataFrame to store the merged data
merged_df = pd.DataFrame()

# Iterate through each sheet and merge the data
for sheet_name in sheet_names:
    # Read data from the current sheet
    df = pd.read_excel(xls, sheet_name)

    # Append the data to the merged DataFrame
    merged_df = merged_df.append(df, ignore_index=True)

# Save the merged data to a new Excel file
merged_df.to_excel('merged_data.xlsx', index=False)
```

```python
import pandas as pd

# Load the Excel file
xls = pd.ExcelFile('final GDP.xlsx')

# Get the sheet names
sheet_names = xls.sheet_names

# Initialize an empty DataFrame to store the merged data
merged_df = pd.DataFrame()

# Iterate through each sheet and merge the data column-wise
for sheet_name in sheet_names:
    # Read data from the current sheet
    df = pd.read_excel(xls, sheet_name)

    # Concatenate the data horizontally
    merged_df = pd.concat([merged_df, df], axis=1)

# Save the merged data to a new Excel file
merged_df.to_excel('merged_data_new.xlsx', index=False)
```
